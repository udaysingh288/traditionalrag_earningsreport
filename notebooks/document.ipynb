```python
import json


# Load JSON manually
with open("../data/insurance_info.json", "r") as f:
    data = json.load(f)

# Convert to LangChain Documents
documents = []
for policy in data["insurance_download"]["policies"]:
    doc = Document(
        page_content=f"{policy['provider']['name']} - {policy['policy_type']}",
        metadata={
            "policy_id": policy["policy_id"],
            "provider": policy["provider"]["name"],
            "policy_type": policy["policy_type"],
            "coverage_type": policy["coverage"]["coverage_type"],
            "status": policy["enrollment_status"],
            "deductible": policy["coverage"]["deductible"]["individual"]
        }
    )
    documents.append(doc)

print(f"Loaded {len(documents)} documents\n")
for doc in documents:
    print(f"Content: {doc.page_content}")
    print(f"Metadata: {doc.metadata}\n")
```
